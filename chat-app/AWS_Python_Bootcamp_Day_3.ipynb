# **Socket Programming**

- What are Sockets?
      Interior endpoints made for sending and receiving data.
      - A single network will have 2 sockets.
      - Sockets are a combination of IP adresses and ports.

- Socket Modules
      - socket.socket() - used to mcreate sockets required for both server and client
      - socket.accept() - used to accept a connection
      - socket.bind() - used to bind specific addresses 
      - socket.close() - used to close the socket
      - socket.connect() - used to connect to remote addresses
      - socket.listen() - enables the server to accept the connection

```
#Server
import socket
s=socket.socket(socket.AF_INET, socket.SOCK_STREAM)

s.bind((socket.gethostname(),1024))

s.listen(5)
while True:
    clt, adr=s.accept()
    print("Connection Etablished to: ", adr)
    clt.send(bytes("This is a message","utf-8"))
    clt.close()
```

```
#Client
import socket
s=socket.socket(socket.AF_INET, socket.SOCK_STREAM)

s.connect((socket.gethostname(), 1024))
msg=s.recv(1000)
print(msg.decode("utf-8"))
```
